```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
import torchvision.models as models
from glob import glob
import os

# Define paths for all three datasets
MVTEC_PATH = "/root/.cache/kagglehub/datasets/v-tec-ad" # Adjust based on your actual load path
VISA_PATH = "/root/.cache/kagglehub/datasets/visa"     # Adjust based on your actual load path
BTAD_PATH = "/root/.cache/kagglehub/datasets/thtuan/btad-beantech-anomaly-detection/versions/1"

def collect_all_images():
    # Gather all 'good' or 'train' images across all domains
    patterns = [
        os.path.join(MVTEC_PATH, "**/train/good/*.png"),
        os.path.join(VISA_PATH, "**/train/good/*.JPG"),
        os.path.join(BTAD_PATH, "**/train/ok/*.png")
    ]

    all_paths = []
    for pattern in patterns:
        all_paths.extend(glob(pattern, recursive=True))

    print(f"Total Unified Industrial Images: {len(all_paths)}")
    return all_paths

class SimCLRDataset(Dataset):
    def __init__(self, image_paths, transform):
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert('RGB')
        # Step 3.1: Generate two augmented versions (Positives)
        view1 = self.transform(img)
        view2 = self.transform(img)
        return view1, view2
```

```python
class ContrastiveEncoder(nn.Module):
    def __init__(self, feature_dim=128):
        super().__init__()
        # Load ResNet-50
        resnet = models.resnet50(weights='IMAGENET1K_V1')
        self.backbone = nn.Sequential(*list(resnet.children())[:-1]) # Features: 2048-dim

        # Step 3.2: Projection Head (MLP)
        self.projector = nn.Sequential(
            nn.Linear(2048, 512),
            nn.ReLU(),
            nn.Linear(512, feature_dim)
        )

    def forward(self, x):
        h = self.backbone(x).flatten(1)
        z = self.projector(h)
        return F.normalize(z, dim=1) # Normalized for cosine similarity
```

```python
def contrastive_loss(z1, z2, temperature=0.5):
    """
    NT-Xent loss: Pull z1, z2 closer, push others away.
    """
    batch_size = z1.shape[0]
    features = torch.cat([z1, z2], dim=0) # [2*B, dim]

    # Cosine Similarity Matrix
    logits = torch.matmul(features, features.T) / temperature

    # Mask out self-similarity
    mask = torch.eye(2 * batch_size, device=z1.device).bool()
    logits = logits.masked_fill(mask, -1e9)

    # Ground truth: target z1[i] is z2[i] (which is at index i + batch_size)
    targets = torch.arange(batch_size, device=z1.device)
    targets = torch.cat([targets + batch_size, targets], dim=0)

    return F.cross_entropy(logits, targets)
```

```python
class SimplePathDataset(torch.utils.data.Dataset):
    def __init__(self, path_list, transform=None, is_test=False):
        self.paths = path_list
        self.transform = transform
        self.is_test = is_test

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        path = self.paths[idx]
        img = Image.open(path).convert('RGB')

        if self.transform:
            img = self.transform(img)

        # For test sets, extract label from path (0 for 'ok', 1 for others)
        label = 0
        if self.is_test:
            label = 1 if "ok" not in path.lower() else 0

        return img, label

# --- RE-INITIALIZE LOADERS ---
# Using the lists we found earlier: source_paths and target_paths
train_ds = SimplePathDataset(source_paths, transform=ContrastiveTransform(base_tsf))
test_ds = SimplePathDataset(target_paths, transform=base_tsf, is_test=True)

loader = DataLoader(train_ds, batch_size=16, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=16, shuffle=False)

print(f"✅ Loaders ready with {len(source_paths)} train and {len(target_paths)} test images.")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import models
from tqdm import tqdm
from sklearn.metrics import roc_auc_score
from sklearn.neighbors import NearestNeighbors
from PIL import Image
import gc

# --- HELPER 1: SimCLR Loss (NT-Xent) ---
def nt_xent_loss(z1, z2, temperature=0.5):
    z = torch.cat([z1, z2], dim=0)
    sim = torch.matmul(z, z.T) / temperature
    n = z1.shape[0]
    labels = torch.cat([torch.arange(n) for _ in range(2)], dim=0).to(z1.device)
    mask = torch.eye(2 * n, device=z1.device).bool()
    sim = sim.masked_fill(mask, -1e9)
    return F.cross_entropy(sim, labels)

# --- HELPER 2: Feature Extraction ---
def extract_features(model, dataloader):
    model.eval()
    all_feats, all_labels = [], []
    device = next(model.parameters()).device
    with torch.no_grad():
        for batch in dataloader:
            # Handle the contrastive (v1, v2) tuple vs standard tensor
            imgs = batch[0][0] if isinstance(batch[0], list) else batch[0]
            labels = batch[1]
            proj, _ = model(imgs.to(device))
            all_feats.append(proj.cpu().numpy())
            all_labels.append(labels.numpy())
    return np.concatenate(all_feats), np.concatenate(all_labels)

# --- 1. MODEL DEFINITION ---
class HybridAnomalyEncoder(nn.Module):
    def __init__(self, feature_dim=128):
        super().__init__()
        resnet = models.resnet50(weights='IMAGENET1K_V1')
        self.encoder = nn.Sequential(*list(resnet.children())[:-2])
        self.projector = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(2048, 512),
            nn.ReLU(),
            nn.Linear(512, feature_dim)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(2048, 512, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(512, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.Upsample(scale_factor=8, mode='bilinear', align_corners=True),
            nn.Conv2d(128, 3, kernel_size=3, padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        feat = self.encoder(x)
        proj = F.normalize(self.projector(feat), dim=1)
        rec = self.decoder(feat)
        return proj, rec

# --- 2. STATISTICAL VALIDATION LOOP ---
# Set device globally
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

all_auc_scores = []
num_runs = 3
epochs = 15

for run in range(num_runs):
    print(f"\n🚀 Starting Run {run+1}/{num_runs}")

    # Re-initialize
    model = HybridAnomalyEncoder().to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    scaler = torch.amp.GradScaler('cuda')

    # Training Loop
    for epoch in range(epochs):
        model.train()
        pbar = tqdm(loader, desc=f"Run {run+1} - Epoch {epoch+1}", leave=False)
        
        # CORRECT UNPACKING: (v1, v2) is a tuple from ContrastiveTransform
        for batch in pbar:
            (v1, v2), _ = batch 
            v1, v2 = v1.to(device), v2.to(device)
            optimizer.zero_grad(set_to_none=True)

            with torch.amp.autocast('cuda'):
                z1, rec1 = model(v1)
                z2, rec2 = model(v2)

                loss_simclr = nt_xent_loss(z1, z2)
                loss_rec = F.mse_loss(rec1, v1) + F.mse_loss(rec2, v2)
                loss = loss_simclr + (5.0 * loss_rec)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            pbar.set_postfix(loss=f"{loss.item():.4f}")

    # Evaluation
    print(f"📊 Evaluating Run {run+1}...")
    src_feat, _ = extract_features(model, loader)
    tgt_feat, tgt_labels = extract_features(model, test_loader)

    nn_module = NearestNeighbors(n_neighbors=5, metric='cosine')
    nn_module.fit(src_feat)
    distances, _ = nn_module.kneighbors(tgt_feat)
    
    # Calculate anomaly score (mean distance to neighbors)
    current_auc = roc_auc_score(tgt_labels, distances.mean(axis=1))
    all_auc_scores.append(current_auc)
    print(f"✅ Run {run+1} AUC: {current_auc:.4f}")

    # CRITICAL: Clean memory between runs
    del model, optimizer, scaler
    gc.collect()
    torch.cuda.empty_cache()

# --- 3. FINAL STATISTICAL RESULTS ---
print("\n" + "="*40)
print(f"📊 Final Results: {np.mean(all_auc_scores):.4f} ± {np.std(all_auc_scores):.4f}")
print("="*40)
```

```python
import torch
import torch.nn as nn

class HybridAnomalyEncoder(nn.Module):
    def __init__(self, latent_dim=128):
        super().__init__()
        # Backbone (Shared)
        self.encoder = torch.hub.load('pytorch/vision:v0.10.0', 'resnet50', pretrained=True)
        self.feature_dim = self.encoder.fc.in_features
        self.encoder.fc = nn.Identity() # Remove classification head
        
        # Method 1: SimCLR Projector (Contrastive)
        self.projector = nn.Sequential(
            nn.Linear(self.feature_dim, 512),
            nn.ReLU(),
            nn.Linear(512, latent_dim)
        )
        
        # Method 1: Spatial Decoder (Reconstruction)
        self.decoder = nn.Sequential(
            nn.Linear(self.feature_dim, 512 * 7 * 7),
            nn.Unflatten(1, (512, 7, 7)),
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1), # 14x14
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1), # 28x28
            nn.ReLU(),
            nn.ConvTranspose2d(128, 3, kernel_size=4, stride=8, padding=0),   # Up to 224x224
            nn.Sigmoid()
        )

    def forward(self, x):
        features = self.encoder(x)
        z = self.projector(features)      # For SimCLR Loss
        rec = self.decoder(features)     # For Reconstruction Loss
        return z, rec
```

```python
from torch.utils.data import DataLoader, WeightedRandomSampler
import numpy as np

def get_balanced_fusion_loader(dataset, batch_size):
    # Extract categories from paths (assuming MVTec/Visa structure)
    paths = dataset.paths
    categories = [p.split('/')[-4] if 'mvtec' in p.lower() else p.split('/')[-3] for p in paths]
    
    # Calculate weights: Weight = 1 / Frequency
    unique_cats, counts = np.unique(categories, return_counts=True)
    weights_map = {cat: 1.0 / (count + 1e-6) for cat, count in zip(unique_cats, counts)}
    
    # Apply weights to every sample
    sample_weights = [weights_map[cat] for cat in categories]
    
    sampler = WeightedRandomSampler(
        weights=torch.DoubleTensor(sample_weights),
        num_samples=len(sample_weights),
        replacement=True
    )
    
    return DataLoader(dataset, batch_size=batch_size, sampler=sampler, num_workers=4)
```
